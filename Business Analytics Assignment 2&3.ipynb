**Q.2 Use simple linear regression to predict CO2 emissions using volume and weight separately, plot their graphs and calculate the r values**

**1) a. Using Volume**

```python
import pandas as pd
df1 = pd.read_csv(r"D:\Pandas Tutorial\Regression_Data.csv")
df1
```

```python
vol = df1['Volume']
X = vol.values
X
```

```python
import matplotlib.pyplot as plt
from scipy import stats

X = df1['Volume']
y = df1['CO2']

slope, intercept, r, p, std_err = stats.linregress(X, y)

plt.scatter(X, y)
plt.show()
```

```python
def myfunction(X):
    return slope*X + intercept

Model = list(map(myfunction, X))
```

```python
plt.scatter(X, y)
plt.plot(X, Model)
plt.ylabel("CO2 Emission")
plt.xlabel("Volume")
```

```python
print(r)
```

**b. Using Weight**

```python
x = df1['Weight']
y = df1['CO2']

slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
model = list(map(myfunction, x))
```

```python
plt.scatter(x,y)
plt.plot(x, model)
plt.ylabel("CO2 Emission")
plt.xlabel("Weight")
```

```python
print(r)
```

**Using Machine Learning to predict the genre by Logistic regression**

```python
import numpy as np
from sklearn import linear_model
```

```python
x1 = df2.iloc[:,:-1].values
y1 = df2.iloc[:,-1].values
y1
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.2, random_state = 0 )
```

```python
#scalling
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x1_train = scale.fit_transform(x1_train)
x1_test = scale.transform(x1_test)
```

```python
#Logistic Regression
from sklearn import linear_model 
logr = linear_model.LogisticRegression()
logr.fit(x1_train, y1_train)
```

```python
#predict test results
y1_predicted = logr.predict(x1_test)
```

```python
y1_predicted
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y1_predicted, y1_test)
```
